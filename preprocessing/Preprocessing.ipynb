```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
movies=data.copy()
```

```python
movies.head()
```

```python
def convert_genres(obj):
    genres = []
    for i in obj:
        genres.append(i['name'])
    return genres
```

```python
movies['genres'] = movies['genres'].apply(convert_genres)
```

```python
def convert_director(obj):
    director = []
    for i in obj:
        if i['job'] == 'Director':
            director.append(i['name'])
            break
    return director
```

```python
movies['director'] = movies['crew'].apply(convert_director)
```

```python
movies.head()
```

```python
def convert_cast(obj):
    cast = []
    counter =0
    for i in obj:
        if counter !=3:
            cast.append(i['name'])
            counter+=1
        else:
            break
    return cast
```

```python
movies['actors'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(5)
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
movies['actor_1_name'] = movies['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
movies['actor_2_name'] = movies['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
movies['actor_3_name'] = movies['cast'].map(lambda x: get_actor3(x))
```
